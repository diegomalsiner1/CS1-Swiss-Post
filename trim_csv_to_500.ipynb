# Trim CSV to First 500 Rows
This notebook loads the processed CSV, keeps only the first 500 rows, and saves the result as a new CSV file.

```python
from pathlib import Path

import pandas as pd
```

```python
input_path = Path("03-PROCESSED-DATA/data_processed.csv")
output_path = Path("03-PROCESSED-DATA/data_processed_first_500.csv")

df = pd.read_csv(input_path)
print(f"Original rows: {len(df)}")
df.head()
```

```python
df_trimmed = df.iloc[:500].copy()
print(f"Trimmed rows: {len(df_trimmed)}")
df_trimmed.tail()
```

```python
df_trimmed.to_csv(output_path, index=False)
print(f"Saved trimmed file to: {output_path}")
```
